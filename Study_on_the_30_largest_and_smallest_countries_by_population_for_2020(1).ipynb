```python
import pandas as pd

file_path = 'Total_Population.csv'
data = pd.read_csv(file_path, skiprows=4)

data.head()
```

```python
import matplotlib.pyplot as plt


data_2020 = data[['Country Name', '2020']].dropna()

non_countries = [
    'World', 'IDA & IBRD total', 'Low & middle income', 'Middle income', 'IBRD only',
    'Early-demographic dividend', 'Lower middle income', 'Upper middle income',
    'East Asia & Pacific', 'Late-demographic dividend', 'East Asia & Pacific (excluding high income)',
    'East Asia & Pacific (IDA & IBRD countries)', 'South Asia (IDA & IBRD)', 'South Asia', 'IDA total',
    'High income', 'OECD members', 'IDA only', 'Sub-Saharan Africa (IDA & IBRD countries)',
    'Sub-Saharan Africa', 'Sub-Saharan Africa (excluding high income)', 'Post-demographic dividend',
    'Least developed countries: UN classification', 'Pre-demographic dividend',
    'Fragile and conflict affected situations', 'Europe & Central Asia',
    'Heavily indebted poor countries (HIPC)', 'Africa Eastern and Southern',
    'Low income', 'Latin America & Caribbean', 'Latin America & the Caribbean (IDA & IBRD coun...)',
    'Latin America & Caribbean (excluding high income)', 'IDA blend',
    'Middle East & North Africa', 'Africa Western and Central', 'Europe & Central Asia (IDA & IBRD countries)',
    'Arab World', 'European Union', 'Middle East & North Africa (excluding high inc...)',
    'Middle East & North Africa (IDA & IBRD countries)', 'North America', 'Euro area',
    'Central Europe and the Baltics'
]

data_2020_filtered = data_2020[~data_2020['Country Name'].isin(non_countries)]

data_2020_sorted_largest = data_2020_filtered.sort_values(by='2020', ascending=False).head(30)

data_2020_sorted_largest




plt.figure(figsize=(12, 8))
plt.barh(data_2020_sorted_largest['Country Name'], data_2020_sorted_largest['2020'], color='green')
plt.xlabel('Population in 2020')
plt.ylabel('Country')
plt.title('Top 30 Countries by Population in 2020')
plt.gca().invert_yaxis()
plt.show()
```

```python
import matplotlib.pyplot as plt

data_2020 = data[['Country Name', '2020']].dropna()

data_2020_sorted_smallest = data_2020.sort_values(by='2020', ascending=True).head(30)

data_2020_sorted_smallest
```
